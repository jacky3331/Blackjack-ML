```python
# Imports
import itertools
import random

import pandas as pd
```

```python
def reset_deck(deck_multiple):
    vals = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'jack', 'queen', 'king', 'ace']
    suits = ['spades', 'clubs', 'hearts', 'diamonds']
    
    deck = list(itertools.product(vals, suits))
    deck = deck * deck_multiple
    random.shuffle(deck)
    
    return deck

def simulate_blackjack(deck_multiple, num_hits):
    deck = reset_deck(deck_multiple)
    dealer = []
    player1 = []
    
    result = None
    bust = False
    
    game_info = {}
    
    dealer.append(deck.pop())
    player1.append(deck.pop())

    dealer.append(deck.pop())
    player1.append(deck.pop())

    for i in range(num_hits):
        player1.append(deck.pop())
        
        if sum([deck_values_1[card] for card in player1]) == 21 or sum([deck_values_11[card] for card in player1]) == 21:
            bust = False
            break
            
        elif sum([deck_values_1[card] for card in player1]) > 21:
            bust = True
            break
    
    if bust == True:
        result = "Lose"
    
    elif bust == False:
        # Dealer is never gonna bust with just 2 cards, just use convert dealer aces to 11
        
    # Dealer has to draw if value of cards less than 17
    
    # There's an easy way to do this. Just create variables for both conversions for both players, and use a bunch of if statements.
    
        if sum([deck_values_11[card] for card in player1]) > sum([deck_values_11[card] for card in dealer]) and sum([deck_values_11[card] for card in player1]) <= 21:
            result = "Win"
        elif sum([deck_values_1[card] for card in player1]) > sum([deck_values_11[card] for card in dealer]):
            result = "Win"
        elif sum([deck_values_1[card] for card in player1]) == sum([deck_values_1[card] for card in dealer]) or sum([deck_values_11[card] for card in player1]) == sum([deck_values_11[card] for card in dealer]):
            result = "Tie"
        elif sum([deck_values_11[card] for card in player1]) == sum([deck_values_1[card] for card in dealer]) or sum([deck_values_1[card] for card in player1]) == sum([deck_values_11[card] for card in dealer]):
            result = "Tie"
        else:
            result = "Lose"

    game_info["player_hand"] = player1
    game_info["dealer_hand"] = dealer
    game_info["bust"] = bust
    game_info["outcome"] = result
#     return tuple([player1] + [dealer] + [bust] + result)
    return game_info
```
